# 크롤링 이란?

크롤링[crawling] : 무수히 많은 컴퓨터에 분산 저장되어 있는 문서를 수집하여 검색 대상의 색인으로 포함시키는 기술. 어느 부류의 기술을 얼마나 빨리 검색 대상에 포함시키냐 하는 것이 우위를 결정하는 요서로서 최근 웹 검색의 중요성에 따라 발전되고 있다.

웹 크롤링 : 웹 크롤링은 콘텐츠를 수집하기 위해 자동으로 웹사이트를 방문하는 프로세스다.

웹 크롤러 : 자동으로 웹 페이지를 방문해 콘텐츠를 가져오고 URL을 추출해 낸다. 웹 크롤러의 다른 이름은 웹 스파이더, 봇 또는 자동화 색인기이다.

```python
import requests
from bs4 import BeautifulSoup

def get_news_title(url):
    response = requests.get(url)
    content = response.text
    
    soup = BeautifulSoup(content)
    title = soup.title.get_text()
    
    return title


print get_news_title('http://auto.daum.net/review/newsview.daum?newsid=MD20161025100010245')

def get_news_title2(url):
    response = requests.get(url)
    content = response.text
    
    soup = BeautifulSoup(content)
    
    h4 = soup.find('h4', attrs = {'class' : 'tit_h4'})
    
    return h4.get_text()


print get_news_title2('http://auto.daum.net/review/newsview.daum?newsid=MD20161025100010245')

    
def get_news_content(url):
    response = requests.get(url)
    content = response.text

    soup = BeautifulSoup(content)
    div = soup.find('div', attrs = {'id' : 'mCenter'})
    if div == None:
        return ''
    
    content = ''
    for paragraph in div.find_all('p'):
        content += paragraph.get_text()
        
    return content.encode('utf-8')

title1 = get_news_title('http://auto.daum.net/review/newsview.daum?newsid=MD20161025100010245')
title2 = get_news_title2('http://auto.daum.net/review/newsview.daum?newsid=MD20161025100010245')
news1 = get_news_content('http://auto.daum.net/review/newsview.daum?newsid=MD20161025100010245')
news2 = get_news_content('http://auto.daum.net/review/newsview.daum?newsid=MD20161025100010245')    

print news1
```

```python
def get_naver_news_title(url):
    response = requests.get(url)
    content = response.text
    soup = BeautifulSoup(content)
    
    h3 = soup.find('h3', attrs = {'id' : 'articleTitle'})
    return h3.get_text()

def get_naver_news_body(url):
    try:
        response = requests.get(url)
        content = response.text
    except:
        return ''
    
    soup = BeautifulSoup(content)
    
    div = soup.find('div', attrs = {'id' : 'articleBodyContents'})
    if div == None:
        return ''
    
    return div.get_text().strip()

print get_naver_news_title('http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=003&aid=0007490800')
print get_naver_news_body('http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=003&aid=0007490800')
```

### 위 코드 해석 

위 코드를 보면

1)  html 태그 (tag), <dt class="articleSubject">와 <dd class="articleSubject">를 발견하실겁니다. 비록 두 태그는 거의 비슷합니다. 비록 첫번째 태그의 원소 (element)는 'dt'이고 두번째 태그의 원소는 'dd'이지만,  어트리뷰트 (attribute)의 Key(?)값과 Value(?)값은 똑같기 때문이지요 
(class= "articleSubject"). 예시로 두개의 태그만 보여드렸지만 다른 뉴스 타이틀들도 class= "articleSubject" 이런 어트리뷰트를 갖고 있지요. 드디어 패턴을 찾았습니다.
  
2) 패턴을 찾았지만 우리가 원하는 것은 class="articleSubject"을 포함한 태그 옆에 있는 또 다른 태그 안에.. 텍스트(빨간색 동그라미)를 가져오는 것입니다.  즉 태그 안에 어트리뷰트인 title 값인 "코스닥 상장기업 글로벌 IR 콘퍼러스' 개최"를 가져와야 됩니다. 

```python
import requests               #Parsing 라이브러리 
from bs4 import BeautifulSoup
url = "https://www.youtube.com/playlist?list=PL1b8E5OIeygBskFhzpfjHYMpyd9tS0SeE"    # 'AOA 노래모음' url주소. 여기서 각 동영상 링크를 가져올 수 있다. 다른 url주소를 가져와도 된다. 

source_code = requests.get(url)
plain_text = source_code.text
soup = BeautifulSoup(plain_text)      # 지정된 웹페이지의 모든 소스코드를 텍스트형식으로 변환, 그리고 'BeautifulSoup'라이브러리에다 저장한다.  
    
for i in soup.findAll('td', attrs={'class' : 'pl-video-title'}): #html에서 반복되는 패턴을 분석해, 'href' (링크주소) 및 'title' (동영상/노래 제목)을 파이썬 콘솔창에서 출력한다. 확인해본 결과 다른 뮤직 재생목록 html의 패턴도 이 것과 똑같다.  
    href = i.next_element.next_element.get('href')
    href = str(href)
    href = href[0:20]
    href = "https://www.youtube.com" + href
    title = i.next_element.next_element.next_element
    print title  # 공백없이 한 줄로 읽히기 위해서 ' end="" ' 라는 명령을 덧붙임. 
    print href
```

# HTTP 리퀘스트 처리하기

* 파이썬에서 http 리퀘스트를 처리하는 방법에는 urllib, urllib2 모듈을 이용하는 방법이 있다.

+ urllib 모듈은 http, ftp, file 등의 처리도 동시에 가능하다.
+ urllib2 모듈은 리퀘스트 시점에 헤더 정보를 입력하는 것이 가능하다. 
+ requests 모듈은 파이썬의 기본 모듈이 아니라 추가적으로 설치하여 사용하는 모듈이다.

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import urllib
import urllib2
import requests

# urllib - Open arbitrary resources by URL
# urllib2 - extensible library for opening URLs -> 파이선3 urllib.request, urllib.error

url = 'http://www.naver.com'

def urllib_sample():
    response = urllib.urlopen(url)
    
    print response.headers
    print response.code
    print response.read()
    
def urllib2_sample():
    opener = urllib2.build_opener()
    opener.addheaders = [("User-agent", "Mozilla/5.0")]
    response = opener.open(url)
    
    print response.headers
    print response.read()

def requests_sample():
    response = requests.get(url, headers={"User-agent" : "Mozilla/5.0"})
    print response.text

    
urllib_sample()
urllib2_sample()
requests_sample()
```
